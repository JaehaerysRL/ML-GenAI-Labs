# LLM Malicious Instruction Defense

本筆記本僅供參考用來計算防禦提示的 token 數，請將你的防禦提示繳交到 JudeBoi 上。    
This notebook is just for your reference to calculate how many tokens your defense prompt contains. Please submit your defense prompt to JudeBoi.

如果想知道你的防禦提示有幾個 token，請執行以下程式碼，並在下方的文字框中輸入你的防禦提示。    
If you want to know the number of tokens of your defense prompt, please run the following code and enter your defense prompt in the textarea below.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("openai/gpt-oss-20b")

label = widgets.HTML(
    value="<b>請在此輸入你的防禦提示：</b><br><b>Please enter your defense prompt here:</b>"
)
text_area = widgets.Textarea(
    value='',
    placeholder='',
    layout=widgets.Layout(width='500px', height='200px')
)
button = widgets.Button(
    description='送出 / Submit',
    button_style='success'
)
output = widgets.Output()

def on_button_click(b):
    with output:
        clear_output(wait=True)
        defense_prompt = text_area.value
        inputs = tokenizer(defense_prompt, return_tensors="pt")
        num_tokens = inputs["input_ids"].shape[1]
        print(f"Number of tokens: {num_tokens}")

        if num_tokens >= 1000:
            display(HTML(
                '<span style="color:red; font-weight:600;">'
                '你的防禦提示過長，請縮短至 1000 tokens 以內。<br />'
                'Your defense prompt is too long. Please shorten it to within 1000 tokens.'
                '</span>'
            ))

button.on_click(on_button_click)

display(label, text_area, button, output)
```
